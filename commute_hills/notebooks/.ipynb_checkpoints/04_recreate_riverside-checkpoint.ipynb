```python
# import altair with an abbreviated alias
import altair as alt
import csv
import pandas as pd
```

<h3>Adjustment 1: Drop bad readings between 136th and 129th</h3>

```python
first_bad_gps = df_walk.loc[df_walk['lat'] == 40.82169].index[0]
first_bad_gps
```

```python
df_walk = pd.read_csv("../results/commute_broadway_stats.csv")
```

```python
last_bad_gps = df_walk.loc[df_walk['lat'] == 40.81676].index[0]
last_bad_gps
```

```python
df_walk = df_walk.drop(df_walk.index[first_bad_gps:last_bad_gps], axis=0)
# for gps in df_walk[68:93]:
#     print(gps.str)
```

```python
#review results
alt.Chart(df_walk).mark_circle(size=30).encode(
    x='distance_sum',
    y='elev',
    #color='slope',
    tooltip=['lat', 'long', 'slope_angle']
).interactive()
```

<h3>Adjustment 2: Append corrected readings between 136th and 129th</h3>

```python
df_walk_bridge = pd.read_csv("../results/commute_riverside_bridge_136_stats.csv")
```

```python
#review the bridge
alt.Chart(df_walk_bridge).mark_circle(size=30).encode(
    x='distance_sum',
    y='elev',
    #color='slope',
    tooltip=['lat', 'long', 'slope_angle']
).interactive()
```

```python
df_walk_riverside = pd.concat([df_walk,df_walk_bridge])
```
