<h1>Find inflection points on Broadway</h1>

```python
df_walk_broadway = pd.read_csv("../results/commute_broadway_stats.csv", index_col=[0])
df_walk_broadway
```

```python
alt.Chart(df_walk_broadway).mark_circle(size=30).encode(
    x='distance_sum',
    y='elev',
    #color='slope',
    #tooltip=['lat', 'long', 'slope_angle']
).interactive()
```

<h3>Observations</h3>
Downward descent begins at Broadway & 139th (lat = 40.82305)<br>
Hill begins at Broadway & 125th (lat = 40.81592)

<h3>Encode Walk Sections Based on Inflections</h3>

```python
#find section endpoints
end_segment1 = df_walk_broadway.loc[df_walk_broadway['lat'] == 40.82305].index[0]
end_segment2 = df_walk_broadway.loc[df_walk_broadway['lat'] == 40.81592].index[0]
end_segment3 = len(df_walk_broadway)
df_walk_broadway["seg_id"]=0
df_walk_broadway["segment"]=0
```

```python
#assign segments
df_walk_broadway["seg_id"].loc[0:end_segment1] = 1
df_walk_broadway["seg_id"].loc[end_segment1:end_segment2] = 2
df_walk_broadway["seg_id"].loc[end_segment2:end_segment3] = 3
df_walk_broadway["segment"].loc[0:end_segment1] = '157_139'
df_walk_broadway["segment"].loc[end_segment1:end_segment2] = '139_125'
df_walk_broadway["segment"].loc[end_segment2:end_segment3] = '125_116'
df_walk_broadway
```

```python
df_walk_broadway.segment.value_counts()
```

```python
alt.Chart(df_walk_broadway).mark_circle(size=30).encode(
    x='distance_sum',
    y='elev',
    color='segment:N',
    #tooltip=['lat', 'long', 'slope_angle']
).interactive()
```

```python
df_walk_broadway.to_csv("../results/commute_broadway_segmented.csv")
```

<h1>Fix Riverside Elevation Data</h1>

<h3>Step 1: Read original GPS dataset to identify bad readings:</h3>
   <strong>
    <ul>
    <li>1st bridge: 155th (over Trinity Cemetery)- drop data
        <li>2nd bridge: between 136th and 129th (over 12th Avenue)- drop data and replace with Google Earth readings
    </ul>
    </strong>

```python
df_walk_riverside = pd.read_csv("../data/commute_riverside.csv", index_col=[0])
```

```python
gps_bridge1_start = df_walk_riverside.loc[df_walk_riverside['lat'] == 40.83404].index[0]
gps_bridge1_start
```

```python
gps_bridge1_end = df_walk_riverside.loc[df_walk_riverside['lat'] == 40.83398].index[0]
gps_bridge1_end
```

```python
gps_bridge2_start = df_walk_riverside.loc[df_walk_riverside['lat'] == 40.82169].index[0]
gps_bridge2_start
```

```python
gps_bridge2_end = df_walk_riverside.loc[df_walk_riverside['lat'] == 40.81676].index[0]
gps_bridge2_end
```

```python
gps_end = len(df_walk_riverside)
gps_end
```

<h3>Step 2 - Drop Data for the 1st bridge</h3>

```python
df_walk_riverside = df_walk_riverside.drop([gps_bridge1_start,gps_bridge1_end+1])
df_walk_riverside
```

<h3>Step 3 - Surgically Replace Data for the 2nd bridge</h3>
Append altitude data on Riverside bridge manually looked up for each block between 129th and 136th on <a href="https://earth.google.com/web/@40.81864262,-73.95820922,7.25365032a,400.670882d,35y,0.88414262h,0t,0r">Google Earth</a>

```python
#for the 2nd bridge replacement, can't simply drop the bad records
#need to track walk in segments
```

```python
df_walk_riverside_start = df_walk_riverside[0:gps_bridge2_start]
df_walk_riverside_start
```

```python
df_walk_riverside_end = df_walk_riverside[gps_bridge2_end+1:gps_end]
df_walk_riverside_end
```

<h3>Adjustment 2</h3> Append altitude data on Riverside bridge manually looked up for each block between 129th and 136th on <a href="https://earth.google.com/web/@40.81864262,-73.95820922,7.25365032a,400.670882d,35y,0.88414262h,0t,0r">Google Earth</a>

```python
#append
df_walk_bridge = pd.read_csv("../data/commute_riverside_bridge_136.csv")
df_walk_riverside_cleaned = pd.concat([df_walk_riverside_start,df_walk_bridge,df_walk_riverside_end])
df_walk_riverside_cleaned
```

```python
#sort, re-index, drop extra columns
df_walk_riverside_cleaned = (df_walk_riverside_cleaned
    .reset_index()
    .drop(columns =["index","cross_street","lat_long_google_maps"])
)
df_walk_riverside_cleaned
```

```python
df_walk_riverside_cleaned.to_csv("../data/commute_riverside_cleaned.csv")
```
