<h1>Import Libraries</h1>

```python
# import altair with an abbreviated alias
import altair as alt
import csv
import pandas as pd
```

<h1>Review Riverside</h1>

```python
df_walk_riverside = pd.read_csv("../results/commute_riverside_cleaned_stats.csv", index_col=[0])
df_walk_riverside
```

```python
alt.Chart(df_walk_riverside).mark_circle(size=30).encode(
    x='distance_sum',
    y='elev',
    #color='segment',
    tooltip=['lat', 'long', 'slope_angle']
).interactive()
```

<h3>segmenting</h3>
try to segment at 139th & 125th to match Broadway data

```python
#find section endpoints
#no wizardry here, just picked segments based on scatterplot...
#...at around 1,500m and 2,500m to match Broadway segmentation
end_segment1 = df_walk_riverside.loc[df_walk_riverside['lat'] == 40.82344].index[0]
end_segment2 = df_walk_riverside.loc[df_walk_riverside['lat'] == 40.81669].index[0]
end_segment3 = len(df_walk_riverside)
df_walk_riverside["seg_id"]=0
df_walk_riverside["segment"]=0
```

```python
#assign segments
df_walk_riverside["seg_id"].loc[0:end_segment1] = 1
df_walk_riverside["seg_id"].loc[end_segment1:end_segment2] = 2
df_walk_riverside["seg_id"].loc[end_segment2:end_segment3] = 3
df_walk_riverside["segment"].loc[0:end_segment1] = '157_139'
df_walk_riverside["segment"].loc[end_segment1:end_segment2] = '139_125'
df_walk_riverside["segment"].loc[end_segment2:end_segment3] = '125_116'
df_walk_riverside
```

```python
alt.Chart(df_walk_riverside).mark_circle(size=30).encode(
    x='distance_sum',
    y='elev',
    color='segment:N',
    #sort = 'seg_id',
    #color = alt.color('segment:N').sort('seg_id'),
    tooltip=['lat', 'long', 'slope_angle']
).interactive()
```

```python
df_walk_riverside.to_csv("../results/commute_riverside_segmented.csv")
```
